```python
import pandas as pd
import numpy as np
import re
from collections import Counter
import nltk
nltk.download('wordnet')
nltk.download('punkt')
from nltk.stem import WordNetLemmatizer 
```

```python
data = pd.read_csv("tweets.csv")
lemma_array = []
lemmatizer = WordNetLemmatizer()
```

```python
new_data = data.replace(to_replace='[^\sa-zA-Z0-9]@', value = "", regex=True)
```

```python
final_data = new_data.replace(to_replace='RT', value="", regex=True)
```

```python
final_data.rename(columns={'screenName':'Handler'}, inplace=True)
final_data['Handler'] = final_data['Handler'].replace(to_replace='[^\sa-zA-Z0-9]', value='', regex=True)
```

```python
final_data['statusSource'] = final_data['statusSource'].replace('<a href="https://mobile.twitter.com" rel="nofollow">Twitter ', value='', regex=True)
final_data['statusSource'] = final_data['statusSource'].replace('<a href="http://twitter.com/download/iphone" rel="nofollow">Twitter for ', value='', regex=True)
final_data['statusSource'] = final_data['statusSource'].replace('<a href="http://twitter.com/download/android" rel="nofollow">Twitter for ', value='', regex=True)
final_data['statusSource'] = final_data['statusSource'].replace('<a href="http://twitter.com" rel="nofollow">Twitter', value='', regex=True)
final_data['statusSource'] = final_data['statusSource'].replace('</a>', value='', regex=True)
```

```python
final_data.rename(columns={'statusSource':'Source'}, inplace=True)
final_data['Handler'] = '@' + final_data['Handler'].astype(str)
```

```python
final_data
```

```python
def corpus_creator(user,data):
    df= data[data['Handler']==user]
    df['Source'] = df['Source'].str.lower()
    df[['retweetCount']] = df['retweetCount']
    out = df[['Handler','retweetCount','Source']]
    return out
```

```python
corpus_creator('@Bartman51', final_data)
```

```python
def ligma(lemma, lemma_array):
    word_list = nltk.word_tokenize(lemma)
    lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in word_list])
    lemma_array.append(lemmatized_output)
    
```

```python
for x in final_data['text']:
    ligma(x, lemma_array)
final_data['lem_text'] = lemma_array
```
